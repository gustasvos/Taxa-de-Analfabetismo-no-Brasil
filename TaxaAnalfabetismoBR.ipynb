```
#Importa a biblioteca pandas
import pandas as pd
```

```
#Lê o arquivo csv e separa as colunas por ';'
filename = 'taxaanalfabetismo.csv'
df = pd.read_csv(filename, sep = ';')
```

```
#Gera o gráfico da taxa de analfabetismo nos anos de 2000 e 2010
df.groupby('Ano').size().sort_values().tail(2).plot(kind='barh', figsize=(10,10))
```

```
#Gera o gráfico com a taxa de analfabetismo por faixa etária
df.groupby('FaixaEtaria').size().sort_values().tail(4).plot(kind='barh', figsize=(10,10))
```

```
#Gera o gráfico dos 10 municípios com maior taxa de analfabetismo
df.groupby('Geografia').size().sort_values().tail(10).plot(kind='barh', figsize=(10,10))
```

```
#Gera o gráfico dos 10 Estados com maior taxa de analfabetismo
df.groupby('UF').size().sort_values().tail(10).plot(kind='barh', figsize=(10,10))
```

```
#Gera o gráfico da taxa de analfabetismo por área
df.groupby('Area').size().sort_values().tail(3).plot(kind='barh', figsize=(10,10))
```

```
#Gera o gráfico da taxa de analfabetismo por sexo
df.groupby('Sexo').size().sort_values().tail(3).plot(kind='barh', figsize=(10,10))
```

```
#Gera um gráfico com a taxa de analfabetismo por cor
df.groupby('Cor').size().sort_values().tail(7).plot(kind='barh', figsize=(10,10))
```
